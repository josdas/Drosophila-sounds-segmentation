```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import numpy as np
from pylab import rcParams
import warnings

warnings.filterwarnings('ignore')
rcParams['figure.figsize'] = 10, 7
```

```python
file_name = '22-03-16-agn-01.wav'
sample_rate, samples = wavfile.read(file_name)
```

```python
# Сглаживание и уменьшение записи
SAMPLES_COUNT = 10**5
samples = signal.savgol_filter(samples[:SAMPLES_COUNT], 7, 1)
```

```python
plt.plot(np.log10(samples))
'Log plot'
```

```python
plt.plot(samples)
```

```python
Pxx, freqs, bins, im = plt.specgram(samples, NFFT=256, Fs=sample_rate)
```

```python
fourier= np.fft.fft(samples, n=512)
plt.plot(fourier, color='#ff7f00')
```
